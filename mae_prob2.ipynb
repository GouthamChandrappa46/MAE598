```python
import numpy as np
import matplotlib.pyplot as plt
x0 = np.linspace(-1, 1, 30)
diff_f_x = np.linspace(-2, 1, 31)
x1 = (1-diff_f_x) ** 3
plt.figure(2)

# plotting the frame of the coordinate
plt.plot(x0, np.zeros(x0.shape[0]), c='yellow', linewidth=0.5)
plt.plot(np.zeros(x1.shape[0]), x1, c='yellow', linewidth=0.5)

# plot constraint
plt.fill_between(diff_f_x, x1, alpha=0.25, label='feasible domain')

# plot point
plt.scatter(1, 0, marker='o', s=40, c='k', label='corner point')

# plot the gradient at conner points
# conner point (0,1)
plt.arrow(0, 1, -0.5, 0, width=0.5, head_width=0.6,
          head_length=0.03, length_includes_head=True)
plt.text(0.5, 1.5, r'$\frac{\partial f}{\partial x}$', fontsize=15)
plt.arrow(0, 1, 0, 4, width=0.05, head_width=0.06,
          head_length=0.3, length_includes_head=True)
plt.text(1.1, 4, r'$\frac{\partial g_1}{\partial x}$', fontsize=15)
plt.arrow(0, 1, 0, -4, width=0.05, head_width=0.06,
          head_length=0.3, length_includes_head=True)
plt.text(1.1, -4, r'$\frac{\partial g_2}{\partial x}$', fontsize=15)

plt.ylim(-5, 30)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend(loc='upper right')
plt.show()
```
